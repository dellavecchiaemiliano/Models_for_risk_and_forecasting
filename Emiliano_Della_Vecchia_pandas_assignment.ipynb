# **Excercise 1**

Downloading data

```python
taxi = "/content/drive/MyDrive/Classroom/Models for Risk and Forecasting/Models for Risk and Forecasting/ASSIGNMENT #1: Pandas/Datasets/nyc_taxi_2019-01.csv"
df = pd.read_csv(taxi)
df
```

1. Create a data frame from the CSV data for January 2019 called `nyc_taxi_2019-01`. Load the CSV file into a data frame, using only the columns `passenger_count`,
`trip_distance`, `payment_type`, and `total_amount`.

```python
df = df[["passenger_count","trip_distance","payment_type","total_amount"]]
df
```

2. How many taxi rides had more than 8 passengers?

```python
df.loc[df['passenger_count']>8,'passenger_count'].count()
```

3. How many taxi rides had zero passengers?

```python
df.loc[df['passenger_count']==0,'passenger_count'].count()
```

4. How many rides cost less than 0?

```python
df.loc[df['total_amount']<0,'total_amount'].count()
```

5. How many taxi rides were paid for in cash, and cost more than $1,000?

```python
expensive_cash = df.loc[(df["payment_type"]==2) & (df["total_amount"]>1000)]
len(expensive_cash)
```

6. How many rides traveled a below-average distance, but cost an above-average amount?

```python
avg_dist = df["trip_distance"].mean()
avg_amount = df["total_amount"].mean()
short_expensive = df.loc[(df["trip_distance"]<avg_dist) & (df["total_amount"]>avg_amount)]
len(short_expensive)
```

1. Create a data frame from the file `nyc-parking-violations-2020.csv`. 
    We are only interested in a handful of the columns:  `Date First Observed,  Plate ID, Registration State, Issue Date,` (a string in MM/DD/YYYY format, always followed by 12:00:00 AM), `Vehicle Make, Street Name, Vehicle Color`

```python
parking = "/content/drive/MyDrive/Classroom/Models for Risk and Forecasting/Models for Risk and Forecasting/ASSIGNMENT #1: Pandas/Datasets/nyc-parking-violations-2020.csv"
columns = ["Date First Observed", "Plate ID", "Registration State", "Issue Date","Vehicle Make", "Street Name", "Vehicle Color"]
df1 = pd.read_csv(parking, usecols=columns)
df1
```

```python
df1["Issue Date"] = pd.to_datetime(df1["Issue Date"], format="%m/%d/%Y %I:%M:%S %p")
df1["Issue Date"] = df1["Issue Date"].dt.strftime("%m/%d")
df1 = df1.sort_values(by="Issue Date")
df1
```

2. Set the data frame’s index to be the `Issue Date` column.


```python
df1 = df1.set_index("Issue Date")
df1
```

3. What were the three most commonly ticket car makes to be issued tickets on January 2nd, 2020?

```python
df2 = df1.loc["01/02"]
df2["Vehicle Make"].value_counts().head(3)
```

4. On what five streets did cars get the most tickets on June 1st, 2020?


```python
df3 = df1.loc["06/01"]
df3["Street Name"].value_counts().head()
```

5. Now set the index to be `Vehicle Color`.


```python
df4 = df1.reset_index()
df4 = df4.set_index("Vehicle Color")
df4
```

6. What were the 10 most common make of vehicles of colour `GREY`?


```python
df5 = df4.loc["GREY"]
df5["Vehicle Make"].value_counts().head(10)
```

7. Repeat step 6. using colors `PINK` and `GREEN`


```python
df6 = df4.loc[["PINK","GREEN"]]
df6["Vehicle Make"].value_counts().head(10)
```

# **Excercise 3**

1. Load the OECD tourism data (from `oecd_tourism.oecd`) into a data frame. We’re
interested in the following columns:

    * `LOCATION`, a three-letter abbreviation for the country name
    * `SUBJECT`, either INT_REC (for tourist funds received) or INT-EXP (for tourist expenses).
    * `TIME`, a year (integer)
    * `Value`, a float indicating thousands of dollars. 

```python
tourism = "/content/drive/MyDrive/Classroom/Models for Risk and Forecasting/Models for Risk and Forecasting/ASSIGNMENT #1: Pandas/Datasets/oecd_tourism.csv"
colnames = ["LOCATION", "SUBJECT", "TIME", "Value"]
tourism_df = pd.read_csv(tourism, usecols=colnames)
tourism_df
```

2. There is a separate CSV file, `oecd_locations.csv`, with two columns. One contains
the three-letter abbreviated location name you saw in the first CSV file. The second is the
full country name. Load this into a data frame and label the two columns as `Country Code` and `Country Full Name` respectively

```python
locations = "/content/drive/MyDrive/Classroom/Models for Risk and Forecasting/Models for Risk and Forecasting/ASSIGNMENT #1: Pandas/Datasets/oecd_locations.csv"
locations_df = pd.read_csv(locations)
```

```python
locations_df.rename(columns={"AUS":"Country Code", "Australia":"Country Full Name"}, inplace=True)
locations_df
```

3. Merge these two data frames together into a new one, In the new data frame, drop
the `LOCATION` column as there will be already a `Country Full Name` column, with the full name of the
country. In the merging procedure, only choose entries common to both the oecd_tourism and oecd_location dataframe (i.e. *inner* join)

```python
merge_df = tourism_df.merge(locations_df, left_on="LOCATION", right_on="Country Code", how="inner")
merge_df = merge_df[["SUBJECT", "TIME", "Value", "Country Code", "Country Full Name"]]
merge_df
```

4. [**OPTIONAL**] Using the merged DataFrame, find the five countries that received the greatest amount of tourist dollars, on average,
accross years in the data set. [hint: use `groupby()`]; however, before grouping, make sure you selected only the countries which *receive* dollars from tourists)

```python
merge_df.loc[merge_df['SUBJECT']=='INT_REC'].groupby('Country Full Name')['Value'].mean().sort_values(ascending=False)
```
